# Описание:
Набор данных содержит 7 461 226 записей ультрамарафонских забегов от 1 641 168 уникальных спортсменов.

Включены следующие столбцы (с типами данных):

1. Год события (int64)
2. Даты события (object)
3. Название события (object)
4. Дистанция/длина события (object)
5. Количество финишировавших участников события (int64)
6. Результаты спортсмена (object)
7. Клуб спортсмена (object)
8. Страна спортсмена (object)
9. Год рождения спортсмена (float64)
10. Пол спортсмена (object)
11. Возрастная категория спортсмена (object)
12. Средняя скорость спортсмена (object)
13. ID спортсмена (int64)


Столбец "Event name" включает информацию о стране, местоположении, которую можно вывести в новый столбец, и аналогично сезонную информацию можно найти в столбце "Event dates" после "Year of event" (их можно извлечь с небольшой обработкой).

Столбец «Event distance/length column» описывает тип гонки, охватывая самые популярные расстояния и длины гонок UM, а также некоторые другие особые модальности (многодневные и т. д.):

Дистанции: 50 км, 100 км, 50 миль, 100 миль
Время: 6 ч, 12 ч, 24 ч, 48 ч, 72 ч, 6 дн., 10 дн.
Кроме того, в других столбцах есть информация о возрасте, поле и скорости (в км/ч).

Получилась выборка с данными о забегах на 50 км и 50 миль, проходивших в США в 2020 году

```python
df2 = df[((df['Event distance/length'].isin(['50mi', '50km'])) & (df['Year of event'] == 2020)) & 
(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')] 
```

```python
df2.head()
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

По данным в столбце "Athlete age category" непонятен точный возраст каждого бегуна, поэтому я создам отдельный столбец "athlete_age", в котором посчитаю возраст

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
df2 = df2.drop(['Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2 = df2.drop(['Athlete club'], axis = 1)
```

```python
df2.head()
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#проверяем наличие дупликатов
```

```python
df2[df2.duplicated()==True]
```

```python
df2.reset_index(drop = True)
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
df2 = df2.rename(columns = {'Year of event' : 'year',
                            'Event dates' : 'race_day',
                            'Event name' : 'race_name',
                            'Event distance/length' : 'race_lenght',
                            'Event number of finishers' : 'race_number_of_finishers',
                            'Athlete performance' : 'athlete_performance',
                            'Athlete country' : 'athlete_country',
                            'Athlete gender' : 'athlete_gender', 
                            'Athlete average speed' : 'athlete_average_speed', 
                            'Athlete ID' : 'athlete_id',})
                            
                            
                            
                            
```

```python
df2.head()
```

```python
df3 = df2[['race_day' , 'race_name', 'race_lenght', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance'
           , 'athlete_average_speed', 'athlete_country']]
```

```python
df3.head()
```

```python
sns.histplot(df3['race_lenght'])
```

Из данного графика можно сделать вывод о том что количество бегунов в забегах на 50 км в США в 2020 году примерно в 4 раза больше количества бегунов в забегах на 50 миль

```python
sns.histplot(df3, x = 'race_lenght', hue = 'athlete_gender')
```

```python
sns.histplot(df3[df3['race_lenght'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = df3, x = 'race_lenght', y = 'athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quart')
```

```python
sns.lmplot(data = df3, x = 'athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender')
```

```python
#Разница в скорости между гонками в 50 км и 50 миль, для мужчин и женщин
```

```python
df3.groupby(['race_lenght', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#Какие возрастные группы лучше пробгают 50 миль (>= 20 забегов)
```

```python
df3.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean',ascending = False).query('count>19')
```

```python
#Какие возрастные группы хуже пробгают 50 миль (>= 10 забегов)
```

```python
df3.query('race_lenght == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean',ascending = True).query('count>9')
```

```python
#В какое время года спортсмены пробегают дистанцию быстрее?
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'winter' if x > 11 else 'fall' if x > 8 
                                             else 'summer' if x > 5 else 'spring' if x > 2 else 'winter')
```

```python
df3.head()
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

Можно сделать вывод о том что летом сложнее всего бегать на большие дистанции, что подтверждает не только наименьшая средняя скорость у бегунов, но и то что летом проводится заметно меньше забегов, чем например осенью и зимой

```python
#только забеги длиной 50 миль
```
